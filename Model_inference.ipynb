# Import Libraries

```python
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import load_model
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

---

# Model Inference

```python
warnings.filterwarnings('ignore')
```

## Load Image

```python
# Function to load, resize and expand image to match the model
def load_resize_image(image_path, target_size=(224, 224)):
    # Load and display the image
    img = load_img(image_path)
    plt.imshow(img)
    plt.axis('off')

    # Resize the image
    img = img.resize(target_size)

    # Convert to array and normalise
    image = img_to_array(img) / 255.0

    # Add batch dimension
    image = np.expand_dims(image, axis=0)
    # print(image.shape)

    return image
```

```python
# Load and display the image
image = load_resize_image('Largelionsmanejellyfish.jpg')
```

## Prediction

```python
def predict(image):
    model = load_model('./deployment/model_4')
    pred = model.predict(image, verbose=0)
    class_labels = ['Barrel Jellyfish', 'Moon Jellyfish', 'Blue Jellyfish', 'Compass Jellyfish', 'Lions Mane Jellyfish', 'Mauve Stinger Jellyfish']
    pred = np.argmax(pred, axis=1)[0]

    print(f'This is a {class_labels[pred]}!')
```

```python
predict(image)
```
